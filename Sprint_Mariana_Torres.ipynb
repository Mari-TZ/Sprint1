```python
# Definicion de Clase Producto
class Producto():
  def __init__(self, id, nombre, descripcion, cantidad, precio):
    self.id = id
    self.nombre = nombre
    self.descripcion = descripcion
    self.cantidad = cantidad
    self.precio = precio

  def __str__(self):
    return f'{self.nombre} - {self.descripcion} - {self.cantidad} - {self.precio}'

# Métodos para obtener y modificar la información de productos
  def get_id(self):
    return self.id

  def set_id(self, nuevo_id):
    self.id = nuevo_id

  def get_nombre(self):
    return self.nombre

  def set_nombre(self, nuevo_nombre):
    self.nombre = nuevo_nombre

  def get_descripcion(self):
    return self.descripcion

  def set_descripcion(self, nuevo_descripcion):
    self.descripcion = nuevo_descripcion

  def get_cantidad(self):
    return self.cantidad

  def set_cantidad(self, nuevo_cantidad):
    self.cantidad = nuevo_cantidad

  def get_precio(self):
    return self.precio

  def set_precio(self, nuevo_precio):
    self.precio = nuevo_precio

# Definicion de Clase Inventario
class Inventario():
  def __init__(self):
    self.datos = dict()

# Definicion de métodos para agregar, eliminar, actualizar, buscar y listar los productos en el inventario

  def agregar_producto(self, producto):
    self.datos[producto.id] = producto

  def eliminar_producto(self, id):
    del self.datos[id]

  def modificar_producto_nombre(self, id, nombre):
    my_producto = self.datos[id]
    my_producto.set_nombre(nombre)

  def modificar_producto_descripcion(self, id, descripcion):
    my_producto = self.datos[id]
    my_producto.set_descripcion(descripcion)

  def modificar_producto_cantidad(self, id, cantidad):
    my_producto = self.datos[id]
    my_producto.set_cantidad(cantidad)

  def modificar_producto_precio(self, id, precio):
    my_producto = self.datos[id]
    my_producto.set_precio(precio)

  def buscar_producto(self, id):
    my_producto = self.datos[id]
    return my_producto

  def listar_productos(self):
    total_productos = 0
    valor_inventario = 0

    for producto in self.datos:
      total_productos += self.datos[producto].cantidad   # total_productos = total_productos + producto.cantidad
      valor_inventario += self.datos[producto].cantidad * self.datos[producto].precio
    return self.datos, total_productos, valor_inventario

# Definición de Menú para usuario
def menu():
    inventario = Inventario()
    while True:
        print("\nMenu:")
        print("1. Agregar producto")
        print("2. Eliminar producto")
        print("3. Actualizar producto")
        print("4. Buscar producto por ID")
        print("5. Listar todos los productos")
        print("6. Salir")



        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            id = input("Ingrese el ID del producto: ")
            nombre = input("Ingrese el nombre del producto: ")
            descripcion = input("Ingrese la descripción del producto: ")
            while True:
              try:
                cantidad = int(input("Ingrese la cantidad del producto: "))
                break
              except:
                print('La cantidad debe ser un numero entero')
            while True:
              try:
                precio = float(input("Ingrese el precio del producto: "))
                break
              except:
                print('El precio debe ser numerico')
            producto = Producto(id, nombre, descripcion, cantidad, precio)
            inventario.agregar_producto(producto)
        elif opcion == "2":
            id = input("Ingrese el ID del producto a eliminar: ")
            try:
              inventario.eliminar_producto(id)
            except:
              print('El producto a eliminar no existe')
        elif opcion == "3":
            id = input("Ingrese el ID del producto a actualizar: ")
            nombre = input("Ingrese el nuevo nombre del producto: ")
            descripcion = input("Ingrese la nueva descripción del producto: ")
            while True:
              try:
                cantidad = int(input("Ingrese la nueva cantidad del producto: "))
                break
              except:
                print('Ingrese una cantidaden número entero')
            while True:
              try:
                precio = float(input("Ingrese el nuevo precio del producto: "))
                break
              except:
                print('El precio debe ser numerico')
            inventario.actualizar_producto(id, nombre, descripcion, cantidad, precio)
        elif opcion == "4":
            id = input("Ingrese el ID del producto a buscar: ")
            producto = inventario.buscar_producto(id)
            if producto:
                print(producto)
            else:
                print("Producto no encontrado.")
        elif opcion == "5":
            resultado = inventario.listar_productos()
            print(f'La cantidad de productos es {resultado[1]}')
            print(f'El valor del inventario es {resultado[2]}')
            for p in resultado[0]:
              print(resultado[0][p])
            print('--')
        elif opcion == "6":
            break
        else:
            print("Opción inválida. Intente nuevamente.")

# Ejecutar el programa
menu()


```
